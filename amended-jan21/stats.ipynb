```python
import pandas as pd
import pylab as pl
from matplotlib.ticker import ScalarFormatter 
import numpy as np
import seaborn as sns

%matplotlib inline
pl.rcParams['figure.figsize'] = (8, 5)

pd.options.display.max_rows = 200
```

```python
#Make floats display as dollar amounts
pd.options.display.float_format = '${:,.2f}'.format
```

```python
#Read in the cleaned data
data = pd.read_csv('cleaned_data.csv', dtype={'zipcode': str})
```

```python
#Check a random sample
data.sample(5)
```

```python
#Plot histogram of claims (log scale)
pl.hist(data['amount'], bins=np.logspace(1, 6, 21))
pl.gca().set_xscale("log")
pl.xlabel('Claim Amount ($)')
pl.ylabel('# of claims')
pl.gca().xaxis.set_major_formatter(ScalarFormatter()) 
pl.ticklabel_format(axis='x', style='plain')
pl.show()
```

```python
#Claims >= $100,000
data[data['amount'] >= 100000].sort_values('amount',ascending=False)
```

```python
#Sum of all claims
data['amount'].sum()
```

```python
#Median claim amount
data['amount'].median()
```

```python
#Number of claims >= $50,000
data['amount'][data['amount'] >= 50000].count()
```

```python
#Number of claims >= $10,000
data['amount'][data['amount'] >= 10000].count()
```

```python
#Top 20 zip codes by # of claims
pd.DataFrame(data['zipcode'].value_counts())[:20]
```

```python
#Top 20 zipcodes by amount
grouped = data.groupby('zipcode')
pd.DataFrame(grouped['amount'].aggregate(np.sum).sort_values(ascending=False))[:20]
```
